# Tratamiento de Datos

```python
# Datos Perdidos
total = train.isnull().sum().sort_values(ascending = False)
percent = ((train.isnull().sum() / train.isnull().count())*100)
tipo = train.dtypes
missing_data = pd.concat([total, percent, tipo], axis = 1, keys = ['Total', '%', 'Tipo'])
missing_data.head(13)
```

**Married**

```python
# 3 nulls
train['Married'].fillna(train['Married'].mode()[0], inplace=True) 
```

**Loan_Amount_Term**

```python
g= sns.catplot(x="Education", y="Loan_Amount_Term", hue="Gender",
            col="Dependents", aspect=0.9,
            kind="swarm", data=train);
```

```python
def impute_Loan_Amount_Term(cols):
    Loan_Amount_Term = cols[0]
    Dependents = cols[1]
    Gender = cols[2]
    Education = cols[3]
    if pd.isnull(Loan_Amount_Term):
        if Dependents == '0' and Gender == 'Male' and Education == 'Graduate':
            return 240
        else:
            return 360
    else:
        return Loan_Amount_Term
```

```python
train['Loan_Amount_Term'] = train[['Loan_Amount_Term','Dependents','Gender','Education']].apply(impute_Loan_Amount_Term,axis=1)
```

**Dependents**

```python
g= sns.catplot(x="Dependents", y="LoanAmount", hue="Married",
            col="Gender", aspect=0.9,
            kind="swarm", data=train);
```

```python
train['Dependents'].fillna(train['Dependents'].mode()[0], inplace=True)
```

**Gender**

```python
g= sns.catplot(x="Dependents", y="Loan_Amount_Term", hue="Gender",
            col="Married", aspect=0.9,
            kind="swarm", data=train);
```

```python
def impute_Gender(cols):
    Gender = cols[0]
    Loan_Amount_Term = cols[1]
    Dependents = cols[2]
    Married = cols[3]
    if pd.isnull(Gender):
        if Loan_Amount_Term == 360 and Dependents in ['0', '1'] and Married == 'No':
            return 'Female'
        else:
            return 'Male'
    else:
        return Gender
```

```python
train['Gender'] = train[['Gender','Loan_Amount_Term','Dependents','Married']].apply(impute_Gender,axis=1)
```

**Self_Employed**

```python
g = sns.catplot(x="Education",col="Gender", hue='Self_Employed',
                data=train, kind="count",
                height=4, aspect=0.7);
```

```python
train['Self_Employed'].fillna(train['Self_Employed'].mode()[0], inplace=True) 
```

**LoanAmount**

```python
sns.relplot(x="LoanAmount", y="ApplicantIncome", data=train);
```

```python
g = sns.catplot(x="LoanAmount", y="Loan_Status",
                kind="box", orient="h", height=1.5, aspect=4,
                data=train.query("LoanAmount > 0"))
g.set(xscale="log");
```

```python
train['LoanAmount'].fillna(train['LoanAmount'].median(), inplace=True)
```

**Credit_History**

```python
g= sns.catplot(x="Credit_History", y="LoanAmount", hue="Gender",
            col="Married", aspect=0.9,
            kind="swarm", data=train);
```

```python
train['Credit_History'].fillna(train['Credit_History'].mode()[0], inplace=True)
```

```python
# Datos Perdidos
total = train.isnull().sum().sort_values(ascending = False)
percent = ((train.isnull().sum() / train.isnull().count())*100)
tipo = train.dtypes
missing_data = pd.concat([total, percent, tipo], axis = 1, keys = ['Total', '%', 'Tipo'])
missing_data.head(13)
```

# Transformacion de variables categoricas

```python
# salvando datos
df_train=train
```

```python
# tratando la variable categroica Gender
df_train.loc[df_train["Gender"] == "Male", "Gender"] = 1
df_train.loc[df_train["Gender"] == "Female", "Gender"] = 0
# tratando la variable categroica Married
df_train.loc[df_train["Married"] == "Yes", "Married"] = 1
df_train.loc[df_train["Married"] == "No", "Married"] = 0
# tratando la variable categroica Education
df_train.loc[df_train["Education"] == "Graduate", "Education"] = 1
df_train.loc[df_train["Education"] == "Not Graduate", "Education"] = 0
# tratando la variable categroica Self_Employed
df_train.loc[df_train["Self_Employed"] == "Yes", "Self_Employed"] = 1
df_train.loc[df_train["Self_Employed"] == "No", "Self_Employed"] = 0
# tratando la variable categroica Property_Area
df_train.loc[df_train["Property_Area"] == "Rural", "Property_Area"] = 1
df_train.loc[df_train["Property_Area"] == "Urban", "Property_Area"] = 2
df_train.loc[df_train["Property_Area"] == "Semiurban", "Property_Area"] = 3
```

```python
df_train["Dependents"].value_counts()
```

```python
# tratando la variable categroica Property_Area
df_train.loc[df_train["Dependents"] == "0", "Dependents"] = 1
df_train.loc[df_train["Dependents"] == "1", "Dependents"] = 2
df_train.loc[df_train["Dependents"] == "2", "Dependents"] = 2
df_train.loc[df_train["Dependents"] == "3+", "Dependents"] = 3
```

```python
# tratando la variable categroica Loan_Status
df_train.loc[df_train["Loan_Status"] == "Y", "Loan_Status"] = 1
df_train.loc[df_train["Loan_Status"] == "N", "Loan_Status"] = 0
```

# Creando nuevas Caracteristicas

```python
#correlacion de datos
corrMatrix = df_train.corr()
sns.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
#Segun la matriz de correlacion la variable "CoapplicantIncome" es la que menos se correlaciona, 
#asi que sus valores se uniran a la variable "ApplicantIncome"
#realizamos la operacion y eliminamos la columna CoapplicantIncome
df_train['ApplicantIncome']=df_train['ApplicantIncome']+df_train['CoapplicantIncome']
df_train.drop(['CoapplicantIncome'], axis=1, inplace=True)
```

# Separando Train y Test

```python
# la variable Loan_ID no tiene sentido para el objetivo
train.drop(['Loan_ID'], axis=1, inplace=True)
```

**convertir de tipos de datos objeto a numerico**

```python
df_train['Gender'] = df_train['Gender'].astype(np.int)
df_train['Married'] = df_train['Married'].astype(np.int)
df_train['Education'] = df_train['Education'].astype(np.int)
df_train['Self_Employed'] = df_train['Self_Employed'].astype(np.int)
df_train['Loan_Status'] = df_train['Loan_Status'].astype(np.int)
df_train['Dependents'] = df_train['Dependents'].astype(np.int)
```

```python
#creamos los parametros para dividir los datos
X = np.array(df_train.drop(['Loan_Status'], 1))
y = np.array(df_train['Loan_Status'])
```

```python
#dividimos los datos
X_train, X_test, y_train, y_test = train_test_split(X, 
                                                    y, 
                                                    test_size=0.20)
print('Son {} datos para entrenamiento y {} datos para prueba'.format(X_train.shape[0], X_test.shape[0]))
```

# Regresin Logstica

```python
#Modelo de Regresin Logstica
reg_Logistica = LogisticRegression() #max_iter=1000, multi_class='ovr', n_jobs=1
reg_Logistica.fit(X_train, y_train)
print('Precisin Regresin Logstica: {}'.format(reg_Logistica.score(X_train, y_train)))
Y_pred = reg_Logistica.predict(X_test)
print('Matriz de confusion\n{}'.format(confusion_matrix(y_test, Y_pred)))
```

# rbol de Clasificacin

```python
#Modelo de rboles de Decisin Clasificacin
arb_decision = DecisionTreeClassifier(criterion='gini', max_depth=10)
arb_decision.fit(X_train, y_train)
Y_pred = arb_decision.predict(X_test)
print('Precisin rboles de Decisin Clasificacin: {}'.format(arb_decision.score(X_train, y_train)))
print('Matriz de confusion\n{}'.format(confusion_matrix(y_test, Y_pred)))
```

# K Vecinos ms Cercanos

```python
#Modelo de K Vecinos ms Cercanos
k_vecinos = KNeighborsClassifier(n_neighbors=5)
k_vecinos.fit(X_train, y_train)
Y_pred = k_vecinos.predict(X_test)
print('Precisin Vecinos ms Cercanos: {}'.format(k_vecinos.score(X_train, y_train)))
print('Matriz de confusion\n{}'.format(confusion_matrix(y_test, Y_pred)))
```

# Maquina de Vectores

```python
#Modelo de Maquina de Vectores
maq_vectores = SVC()
maq_vectores.fit(X_train, y_train)
Y_pred = maq_vectores.predict(X_test)
print('Precisin Mquinas de Vectores de Soporte: {}'.format(maq_vectores.score(X_train, y_train)))
print('Matriz de confusion\n{}'.format(confusion_matrix(y_test, Y_pred)))
```
